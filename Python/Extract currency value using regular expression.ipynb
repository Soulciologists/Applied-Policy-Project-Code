```python
combined = pd.DataFrame()
directory = "Finished"
files = os.listdir(directory)

for xls in files:
    if xls.endswith(".xlsx"):
        file_path = os.path.join(directory, xls)
        tbc = pd.read_excel(file_path)
        investor_name = os.path.splitext(xls)[0]
        tbc['Investor'] = investor_name        
        tbc = tbc.reindex(columns=['Investor'] + [col for col in tbc.columns if col != 'Investor'])        
        combined = pd.concat([combined, tbc])

```

```python
import re

pattern = r'(\d*\.?\d+)([KMBT]?)'
numeric_values = combined['Money Raised'].apply(lambda x: re.findall(pattern, str(x)))
print(numeric_values)
```

```python
def process_numeric_values(matches):
    if matches:
        value, suffix = matches[0]
        if suffix == 'K':
            return float(value) * 1e3
        elif suffix == 'M':
            return float(value) * 1e6
        elif suffix == 'B':
            return float(value) * 1e9
        elif suffix == 'T':
            return float(value) * 1e12
        else:
            return float(value)
    else:
        return None
    
numeric_values_processed = numeric_values.apply(process_numeric_values)
print(numeric_values_processed)
```

```python
remaining_elements = combined['Money Raised'].apply(lambda x: re.sub(pattern, '', str(x)))
print(remaining_elements)
```

```python
combined["Money Raised (Currency Type)"] = remaining_elements
combined['Money Raised (Values)'] = numeric_values_processed
combined
```
